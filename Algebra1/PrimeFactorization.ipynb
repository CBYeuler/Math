### Prime Factorization

Before we get into the implementation, let's briefly review what a prime number is.

**Prime Number**

A prime number is a natural number greater than 1 that is divisible only by **1** and **itself**.

Mathematically, a number $p$ is prime if its only positive divisors are:

$$
1 \quad \text{and} \quad p
$$

**Examples**

$$
2,\ 3,\ 5,\ 7,\ 11,\ 13,\ 17,\ 19,\ \ldots,\ 43
$$

The number $2$ is a special case because it is the only even prime number. Every other even number is divisible by $2$ and therefore cannot be prime.

Prime numbers are the building blocks of the integers because every positive integer greater than $1$ can be expressed as a product of prime numbers. This process is known as **prime factorization**.


So the idea beginh Prime Factorization is:

Breaking down a number into prime **multiplicands**

$$
8 = 2 \times 2 \times 2\
$$
$$
25 = 5 \times 5\
$$
$$
48 = 2\times 2\times 2\times 2 \times 3\
$$

and here is how it works in `pyhon`:


````python
import sympy as sym
number = 48

sym.factorint(number)#this is the function used in prime factorization
````

Notice that the output of the code above looks very similar to a dictionary - and that is because it *is* a dictionary.

The key-value pairs represent the prime factorization in the following format:

```python
{prime_factor: exponent}
```

or, more generally,

```python
{multiplicand: amount}
```

where:

* **Key** → the prime factor (multiplicand)
* **Value** → the number of times that factor appears in the factorization

For example, the factorization of $360$ is

$$
360 = 2^3 \times 3^2 \times 5^1
$$

which can be represented as:

```python
{2: 3, 3: 2, 5: 1}
```

This representation is particularly useful because it stores all the information needed to reconstruct the original number while remaining compact and easy to manipulate programmatically.


````python
sym.factorint(360)
````

### Challenge: Find Prime Numbers from 0 to 100

Using the function we saw earlier, write a program that loops through all integers from `0` to `100` and prints only the prime numbers.

**Hint**

A prime number has exactly **one prime factor** (excluding `1`).

For example:

```python
{2: 1}      # Prime
{3: 1}      # Prime
{5: 1}      # Prime
{2: 2}      # Not prime (4 = 2²)
{2: 1, 3: 1} # Not prime (6 = 2 × 3)
```



Try solving the problem yourself before looking at the solution.


````python
import sympy as sym
def Primelist(ran):
    for i in range(ran):
        if i < 2:
            print(f"{i} is not a prime number")
            continue
        num = sym.factorint(i)
        if len(num) > 1:
            print(f"{i} is not a prime number")
        elif list(num.values())[0] > 1:
            print(f"{i} is not a prime number")
        else:
            print(f"{i} is a prime number")

Primelist(100)
            

````

````python
#Now of course you can use something like the following to get a faster and easier solution:
import sympy as sym

def Primelist(ran):
    for i in range(ran):
        if sym.isprime(i):
            print(f"{i} is a prime number")
        else:
            print(f"{i} is not a prime number")

Primelist(50)
````

### Listing Composite Numbers and Their Prime Factors

If we wanted to list the numbers that are **not prime** (that is, the composite numbers) and display their prime factors, we could use the factorization dictionary directly.

Recall that a composite number is a positive integer greater than (1) that can be expressed as a product of two or more smaller positive integers.

For example:

$$
12 = 2^2 \times 3^1
$$

which is represented by our function as:

```python
{2: 2, 3: 1}
```

We can iterate through a range of numbers, identify the composite ones, and print both the number and its prime factorization:




````python
import sympy as sym

def Primelist(ran):
    for i in range(ran):
        if i < 2:
            print(f"{i} is neither prime nor composite")
            continue
            
        num = sym.factorint(i)
        
        # If it's not a prime, it's a composite
        if len(num) > 1 or list(num.values())[0] > 1:
            # Formats as 'basis^exponent' if exponent > 1, otherwise just 'basis'
            factors_with_exp = [f"{base}^{exp}" if exp > 1 else str(base) for base, exp in num.items()]
            factors_str = " * ".join(factors_with_exp)
            
            print(f"{i} is a composite number with prime factors of {factors_str}")
        else:
            print(f"{i} is a prime number")

Primelist(50)
````

This representation is particularly useful because it not only tells us that a number is composite, but also shows exactly how it can be decomposed into its prime factors.
