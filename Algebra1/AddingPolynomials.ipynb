### Adding Polynomials

In this section, we will focus on:

* What polynomials are
* How to work with polynomials in Python
* How to add and subtract polynomials

A **polynomial** is an algebraic expression made up of variables and coefficients, where each variable is raised to a non-negative integer power.

A general polynomial can be written as:

$$
a_0 + a_1x + a_2x^2 + a_3x^3 + \dots + a_nx^n
$$

where:

* (a_0, a_1, a_2, \dots, a_n) are called **coefficients**
* (x) is the variable
* (n) represents the highest degree of the polynomial

For example:

$$
3 + 5x + 2x^2 + 7x^3
$$

Here:

* (3) is the constant coefficient $$((a_0))$$
* (5) is the coefficient of $$(x) ((a_1))$$
* (2) is the coefficient of $$(x^2) ((a_2))$$
* (7) is the coefficient of $$(x^3) ((a_3))$$

The highest power of (x) determines the **degree** of the polynomial. In this example, the degree is (3).

$$3 - 5x + 5\frac6x^2 + 12.4x^3 + 2x + x^3$$
$$=$$
$$3 - 3x+ 5\frac 6x^2 + 13.4x^3$$

```python
import sympy as sym
from IPython.display import display, Math
from sympy.abc import x

p1 = 2*x**3 + x**2 - x
p2 = x**3 - x**4 - 4*x**2
display(Math('(%s) + (%s) = %s' %(sym.latex(p1),sym.latex(p2),sym.latex(p1+p2))))
```

```python
p1 = sym.Poly(2*x**3 + x**2 - x)

p1.coeffs()
```

Here we use sym.Poly() to convert a normal algebraic expression into a polynomial object that SymPy can work with.

The polynomial is:

$$
p_1 = 2x^3 + x^2 - x
$$

The terms of this polynomial are:

$$
2x^3 + 1x^2 - 1x
$$

Notice that the coefficient of each term is:

(2) for (x^3)
(1) for (x^2)
(-1) for (x^1)

We can access these coefficients using:

p1.coeffs()

which returns:

[2, 1, -1]

The .coeffs() method gives us a list containing all coefficients of the polynomial, ordered from the highest degree term to the lowest.

In general, for a polynomial:

$$
a_0 + a_1x + a_2x^2 + \dots + a_nx^n
$$

coeffs() returns:

$$
[a_n, a_{n-1}, \dots, a_1, a_0]
$$

so the order follows the degree of the terms rather than the way we originally wrote the expression.

### Exercise
$$x^2 + 2x$$
$$-x^3 + 4x$$
$$x^5 - x^4 + \frac x 4 + 4$$

- for each polynomial if the degree is even, sum the coefficients
- if the degree is odd, count the coefficients

```python
import sympy as sym
from IPython.display import display, Math
from sympy.abc import x

def polyAnalyzer(poly):
    p1 = sym.Poly(poly)
    if p1.degree()%2 == 0:
        print(sum(p1.coeffs()))
    else:
        print(len(p1.coeffs()))

p1=x**2 + 2*x
p2= -x**3 + 4*x
p3=x**5 - x**4 + x/4 + 4

polyAnalyzer(p1)
polyAnalyzer(p2)
polyAnalyzer(p3)
```
