### Solving Inequalities

In this section, we will focus on:

* How to write inequalities in Python
* Solving inequalities using `sympy.solve`
* Representing infinity in SymPy

An **inequality** is a mathematical expression that compares two quantities using symbols such as:

$$

> , \quad <, \quad \geq, \quad \leq
> $$

For example:

$$
4x > 8
$$

Unlike an equation, which asks for specific values that make both sides equal, an inequality asks for the **range of values** that make the statement true.

In the example above, we want to determine all values of (x) for which

$$
4x > 8
$$

holds true.

To solve the inequality, divide both sides by (4):

$$
x > 2
$$

This tells us that **any value greater than (2)** satisfies the inequality.

Some examples of valid solutions are:
$$
3,\ 2.5,\ 100,\ \pi
$$

while values such as

$$
2,\ 1,\ -5
$$

do not satisfy the inequality.

Not all inequalities have a single continuous range of solutions. Consider the following example:

$$
(x-1)(x+3) > 0
$$

The expression is positive when both factors are positive or when both factors are negative.

The critical points are:

$$
x = -3 \quad \text{and} \quad x = 1
$$

Testing the intervals shows that the inequality is satisfied when

$$
x < -3
$$

or

$$
x > 1
$$

Using interval notation, we can write the solution as

$$
(-\infty, -3) \cup (1, \infty)
$$

or equivalently,

$$
(x < -3) \cup (x > 1)
$$

Notice that the solution consists of two separate intervals rather than a single range of values.

In the following sections, we will learn how to represent and solve inequalities symbolically using SymPy.

```python
import sympy as sym
from IPython.display import display, Math

x = sym.symbols('x')

expr = 4*x > 8
sym.solve(expr)
```

The code tells us that the solution set of the inequality is

$$
x > 2
$$

which means that any real number greater than (2) satisfies the inequality.

Using interval notation, we can write the same solution as

$$
(2,\infty)
$$

where $$(\infty)$$ represents positive infinity. Notice that infinity is not a number; rather, it indicates that the interval extends without bound.
