## Linear Equations in Slope-Intercept Form


a slope intercept form essentially looks like :
$ y = mx + b $

"m" being the slope itself
"b" is intercept where the line intercepts the "y" axis

````python
import matplotlib.pyplot as plt

x = [-5,5]
m = 2
b = 1
y = []
for i in range(0, len(x)):
    y.append(m*x[i] + b)

plt.plot(x,y, label= 'y = %sx + %s' %(m,b))
plt.axis("square")
plt.grid()
plt.xlim(x)
plt.ylim(x)

axis = plt.gca()

plt.plot(axis.get_xlim(), [0,0], 'k--')
plt.plot([0,0], axis.get_ylim(), 'k--')
plt.legend()
print(x,y)
````

## Numpy way of solving this problem


````python
import numpy as np
import matplotlib.pyplot as plt

y = m*np.array(x) + b
plt.plot(x,y, label= 'y = %sx + %s' %(m,b))
plt.axis("square")
plt.grid()
plt.xlim(x)
plt.ylim(x)

axis = plt.gca()

plt.plot(axis.get_xlim(), [0,0], 'k--')
plt.plot([0,0], axis.get_ylim(), 'k--')
plt.legend()
````

#### Exercise

$y = 0.7x + -2$ , 
$y = -1.25x + 0.75$

````python
import numpy as np
import matplotlib.pyplot as plt

def plotter(m,b):
    y = m*np.array(x) + b
    plt.plot(x,y, label= 'y = %sx + %s' %(m,b))

    plt.xlim(x)
    plt.ylim(x)
    return x, y

plotter(.7, -2)
plotter(-1.25, .75)

axis = plt.gca()

plt.plot(axis.get_xlim(), [0,0], 'k--')
plt.plot([0,0], axis.get_ylim(), 'k--')
plt.legend()
plt.grid()
plt.show()



#y2 = m*np.array(x) + b
````
