$$ci,j = -1 ^ (i+j)$$

```python
import numpy as np
import matplotlib.pyplot as plt

A = np.zeros((10,10))

for i in range(0,np.shape(A)[0]):
    for j in range(0, np.shape(A)[1]):
        A[i,j] = (-1)**(i+j)
print(A)
plt.set_cmap('grey')
plt.tick_params(labelleft=False, labelbottom = False)
plt.imshow(A)
```
