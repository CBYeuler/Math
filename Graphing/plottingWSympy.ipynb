### Plotting with Sympy

```python
import sympy as sym
from sympy.abc import x
y = x**2

sym.plotting.plot(y)
```

```python
import sympy as sym
from sympy.abc import x
y = x**2

p = sym.plotting.plot(y)
print(p)
```

```python
import sympy as sym
from sympy.abc import x
y = x**2

p = sym.plotting.plot(y, show=False)
p.xlim = [0,20]
p.show()
print(p)
```
