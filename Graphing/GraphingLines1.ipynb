## Start/End Notation

the point of this page is to see how to create continues lines in the plotted graph

you will need two points to draw a line i npython as in math 

$Px = (x,y)$
$P1 = (-3, -1)$
$P2 = (4, 4)$

A line could be drawn between P1 and P2 

```python
import matplotlib.pyplot as plt

p1 = [-3, -1]
p2 = [4, 4]

plt.plot(p1, p2)
plt.show()
```

now above is technically is the correct code but as you can see it does not do what we think it should've
the correct code is as follows:

```python
p1 = [-3, -1]
p2 = [4, 4]

plt.plot([p1[0], p2[0]],[p1[1], p2[1]])
plt.show()
```

the correct notation sa you can see is the :
`plt.plot([x coordinates of two points],[y coordinates of the two points])`


```python
p1 = [-3, -1]
p2 = [4, 4]

plt.plot([p1[0], p2[0]],[p1[1], p2[1]], 'r', linewidth = 3)
plt.show()
```
