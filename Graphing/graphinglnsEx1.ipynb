## Graphing Lines Ex1

$$y = (|x|)^(1/2)
x C {-20,-19, ... ,19, 20}$$

```python
import matplotlib.pyplot as plt
#y = abs(x)**(1/2)
x = range(-20, 21)

for i in x:
    y = abs(i)**(1/2)
    plt.plot([i,y],[0, y])
    
plt.show()
```

```python
import matplotlib.pyplot as plt

x_values = range(-20, 21)
y_values = []

for x in x_values:
    y = abs(x)**(1/2)
    y_values.append(y)
plt.plot(x_values, y_values, 'o-')  


plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)

plt.show()
```
