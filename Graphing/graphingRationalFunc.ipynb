## Graphing Rational Functions

a ratioanal function comes from ratio so its essentially a ==*ratio of polynomials*==
$(x^3+x^2+1)/
(-x^3+x+-1)$ OR $1+(x)^(1/2)$

the way to graphing a ratioanla function or a polynomial in python is to essentially evaluating all the points and connecting the points in the plot

---
$y = 2 - x^2$ :

```python
import numpy as np
import matplotlib.pyplot as plt

x = range(-3,4)

y = np.zeros(len(x))
for i in range(0, len(x)):
    y[i] = 2 - x[i]**2

plt.plot(x, y, 's-')
```

```python
x = np.linspace(-3,3, 10)
y = 2 + np.sqrt(abs(x))
print(x)
plt.plot(x, y, 'm-')
```

- Since the value of y is never "0" it actually never touches the x axis

```python
x = np.linspace(-3,3, 50)#(lower bound, higher bound, steps aka resolution)
y = 2 + np.sqrt(abs(x))
print(x)
plt.plot(x, y, 'm-')
```

- notice that its essentailly the same function but with more points to efine the graph and more resolution

Now let's do it for the first example again:


```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-3,3, 500)

y = np.zeros(len(x))
for i in range(0, len(x)):
    y[i] = 2 - x[i]**2

plt.plot(x, y, 's-')
```

So the graphs we generate do not really show the function itself but rather the points gathered fro mthose functions visualized on the graoh, meaning more points = closer to the reality.

#### Exercise

$$y = x^-1,
y = x^0,
y = x^1,
y = x^2,
y = x^3$$


```python
import numpy as np
import matplotlib.pyplot as plt
x = np.linspace(-4, 4, 100)

for i in range(-1,4):
    y = x**i
    plt.plot(x, y, '-', label=fr'$y = x^{{{i}}}$', linewidth = 3)
print(len(x))
plt.xlabel('x')
plt.ylabel('y')
plt.ylim(-20,20)
plt.xlim(x[0],x[-1])
plt.legend()
plt.grid()


```
