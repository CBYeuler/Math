$$f(x) = -x^2 $$

```python
import matplotlib.pyplot as plt
from matplotlib.patches import Polygon

import numpy as np

x = np.linspace(-2,2, 200)
y = []
for i in range(0, len(x)):
    y.append([x[i],-x[i]**2])

y = np.array(y)
p = Polygon(y, alpha = .3)
fig,ax = plt.subplots()
ax.add_patch(p)

ax.set_ylim([-4,4])
ax.set_xlim([-4,4])
plt.show()
```

#### First Solution:

```python
import matplotlib.pyplot as plt
from matplotlib.patches import Polygon

import numpy as np

x = np.linspace(-2,2, 200)
y = []
for i in range(0, len(x)):
    y.append([x[i],-x[i]**2])

y = np.array(y)

y2 = np.array([[-.5, -4],[-.5,-2.5],[.5, -2.5],[.5,-4]])
p = Polygon(y, alpha = 1, facecolor = 'y', edgecolor = 'k')
p2 = Polygon(y2, alpha = 1, facecolor = 'k')
fig,ax = plt.subplots()
ax.add_patch(p)
ax.add_patch(p2)


ax.set_ylim([-4,0.5])
ax.set_xlim([-2,2])
plt.show()
```

#### Second solution:

```python
x = np.linspace(-2,2,200)
f = -x**2

pnts = np.vstack((x,f)).T

p = Polygon(pnts, facecolor = 'g', alpha = .2)

fig,ax = plt.subplots()
ax.add_patch(p)

plt.plot(x,f,'k')
plt.show()
```
