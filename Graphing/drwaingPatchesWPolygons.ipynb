## Drawing Patches With Polygons

```python
import matplotlib.pyplot as plt
import numpy as np

from matplotlib.patches import Polygon

y = np.array([ [1,1],[2,3],[3,1] ])
```

so the input into the polygon function always needs to be some number of points.
so each row is a point and than the columns correspond to the x locations and the y locations

```python
y
```

```python
p = Polygon(y, alpha = .3)

fig,ax = plt.subplots()

ax.add_patch(p)

ax.set_ylim([0,4])
ax.set_xlim([0,4])

plt.show()
```

```python
y = np.array([ [1,1],[2,3],[3,1] ])
p = Polygon(y, alpha = .3)

y2 = np.array([ [2,2],[3,4],[4,2] ])
p2 = Polygon(y2,facecolor='m', alpha = .5)

fig,ax = plt.subplots()

ax.add_patch(p)
ax.add_patch(p2)

ax.set_ylim([0,4])
ax.set_xlim([0,4])

plt.show()
```

```python
y = np.array([ [1,1],[2,3],[3,1] ])
p = Polygon(y, alpha = .3)

y2 = np.array([ [2,2],[2.5,4],[3.5,1] ])
p2 = Polygon(y2,facecolor='m', alpha = .5, edgecolor = 'k')

fig,ax = plt.subplots()

ax.add_patch(p)
ax.add_patch(p2)

ax.set_ylim([0,4])
ax.set_xlim([0,4])

plt.show()
```
