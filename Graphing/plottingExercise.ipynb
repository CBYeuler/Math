## Plotting Coordinates on a Plane Exercises

$$y = x^2 - 3x$$

$$x C {-10,-9, ... , 10, 11}$$

```python
import matplotlib.pyplot as plt
import sympy as sym

x = range(-10, 11)
axis = plt.gca()

for i in x:
    y = i**2 - 3*i
    plt.plot(i, y, 'o')
x = sym.symbols('x')
y = x**2 - 3*x
plt.xlabel('x')
plt.ylabel('f(x) = %s' %y)
axis.set_xlim([-11, 11])
plt.grid(True)
plt.show()
```
