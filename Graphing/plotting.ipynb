### Plotting coordinates on a plane

```python
import matplotlib.pyplot as plt

x = 3
y = 5

plt.plot(x,y, 'ro')
plt.show()
```

```python
x = 3
y = 5

plt.plot(x,y, 'ro')
plt.axis('square') # declares the shapoe of the graph
plt.axis([-6,6, -6, 6])
plt.grid()#adds grid to the graph
plt.show()
```

```python
x = [-4 , 5 ,2 , 4, 5, 6 ,3]
y = [0, 3, 4, 5, 2, -3, 4] 

for i in range(0, len(x)):
    plt.plot(x[i], y[i], 'o', label=x[i]) #label adds in the label as a legend o nthe graph 
    plt.axis('square')
    plt.grid()

plt.legend()
plt.show()
```

```python
x = [-4 , 5 ,2 , 4, 5, 6 ,3]
y = [0, 3, 4, 5, 2, -3, 4] 

for i in range(0, len(x)):
    plt.plot(x[i], y[i], 'o', label='point %s' %i) #label adds in the label as a legend o nthe graph 
    plt.axis('square')
    plt.grid()

plt.legend()
plt.show()
```

```python
plt.plot(4,2, 'r+')

axis = plt.gca()

ylim = axis.get_ylim()
print(ylim)

axis.set_ylim([ylim[0], 7.3476])

plt.show()
```
