### Summation and Products

In this section, we will focus on:

* A reminder on lists in Python
* Learning how to compute sums and products
* Understanding what a cumulative sum means

In Python, we can store multiple values inside a list:

```python
a = [1, 3, 4, 1, 6]
```

Imagine we have the following sequence of numbers:

$$
a = {1,3,4,1,6}
$$

If we want to describe the sum of all numbers in this sequence in a generalized way, without being tied to these specific values, we can write:

$$
a_1 + a_2 + a_3 + \dots + a_n
$$

where each $$(a_i)$$ represents an element of the sequence and (n) represents the total number of elements.

Instead of writing every term manually, we can use the **summation symbol**:

$$
\sum_{i=1}^{n} a_i
$$

This means:

"Add every element $$(a_i)$$ starting from $$(i=1)$$ until $$(i=n)$$."

For example, for our sequence:

$$
a = {1,3,4,1,6}
$$

the summation would be:

$$
\sum_{i=1}^{5} a_i = 1+3+4+1+6
$$

which gives:

$$
15
$$

As we can see, the summation symbol allows us to represent the addition of all numbers within a given range.

But what if we wanted to represent the **multiplication** of all numbers instead?

For multiplication, we use a similar notation called the **product symbol**:

$$
\prod_{i=1}^{n} a_i
$$

This means:

"Multiply every element (a_i) starting from (i=1) until (i=n)."

For example, using the same sequence:

$$
a = {1,3,4,1,6}
$$

the product would be:

$$
\prod_{i=1}^{5} a_i = 1 \times 3 \times 4 \times 1 \times 6
$$

which gives:

$$
72
$$

Just like summation can be written as:

$$
\sum_{i=1}^{n} a_i
$$

a product can be written as:

$$
\prod_{i=1}^{n} a_i
$$

The difference is:

* (\sum) → add all elements
* (\prod) → multiply all elements


### Cumulative Sum

The next concept we want to focus on is the **cumulative sum**.

A cumulative sum is similar to a normal summation, except instead of calculating only the final total, we keep track of the sum as it grows.

Imagine we have a sequence:

$$
A = {1,3,4,1,6}
$$

The normal summation gives us:

$$
\sum_{i=1}^{5} A_i = 1+3+4+1+6 = 15
$$

However, the cumulative sum gives us a new sequence where each element contains the sum of all previous elements up to that position.

If we call the cumulative sum sequence (C), then the (k)-th element is:

$$
C_k = \sum_{i=1}^{k} A_i
$$

This means that each element of (C) contains the sum of the first (k) elements of (A).

$$C_1 = a_1$$
$$C_2 = a_1 + a_2 $$
$$C_3 = a_1 + a_2 + a_3 $$
$$C_k = a_1 + a_2 + .... + a_k $$

For our example:

$$
A = {1,3,4,1,6}
$$

the cumulative sum becomes:

$$
C = {1,\ 1+3,\ 1+3+4,\ 1+3+4+1,\ 1+3+4+1+6}
$$

which gives:

$$
C = {1,\ 4,\ 8,\ 9,\ 15}
$$

In Python, this concept is commonly implemented using cumulative sum functions such as `numpy.cumsum()`.



````python
import numpy as np
import matplotlib.pyplot as plt

lst = [1,3,4,1,6]

sum(lst)
````

````python
import numpy as np

lst = [1,3,4,1,6]

np.sum(lst)
````

````python
#product of the list
np.prod(lst)
````

````python
#cumilitive sum of the list
np.cumsum(lst)
````

````python
np.cumsum(lst)
plt.plot(lst, 'rs-', label = 'list')
plt.plot(np.cumsum(lst),'bo-', label = 'cumilitive sum')
plt.legend()
plt.show()
````
